```python
movie_plots_file = 'movie_plots.csv'

import pandas as pd
import numpy as np

# Read in the movie plots
df = pd.read_csv(movie_plots_file)
print(df.shape)
df.head()
```

```python
batmanbegins_df = df[df['title'].str.match('Batman Begins')]
batmanbegins_plot = batmanbegins_df.iloc[0]['plot']
batmanbegins_plot
```

```python
import spacy

from spacy import displacy
nlp = spacy.load("en_core_web_trf")
```

```python
doc = nlp(batmanbegins_plot)
```

```python
displacy.render(doc, style='ent', jupyter=True)
```

```python
persons = []
locations = []

for ent in doc.ents:
    if ent.label_ == 'PERSON':
        persons.append(ent.text)
    if ent.label_ == 'GPE':
        locations.append(ent.text)
    
persons = list(set(persons))
locations = list(set(locations))

print(persons)
print(locations)
```

```python
import gensim.downloader as api

model = api.load("word2vec-google-news-300")
```

```python
model['cat'].shape
```

```python
model['Los_Angeles'].shape
```

```python
# king - man = expect monarch or royalty. Didn't get expected answer.
model.most_similar(positive=['king'], negative=['man'], topn=10)
```

```python
# king - man + woman = expect queen.
model.most_similar(positive=['woman', 'king'], negative=['man'], topn=10)
```

```python
# cat - kitten + dog = expect puppy.
model.most_similar(positive=['dog', 'kitten'], negative=['cat'], topn=10)
```

```python
# idli - India + USA = Blueberry pancakes.
model.most_similar(positive=['USA', 'idli'], negative=['India'], topn=10)
```

```python
from_country = 'USA'
to_country = 'India'

names = []
similarnames_list = []

for name in persons:
    similarnames = []

    try:
        similarnames = model.most_similar(positive=[to_country, name], negative=[from_country], topn=10)
        similarnames = [choice[0].replace('_', ' ') for choice in similarnames]
    except:
        pass

    names.append(name)
    similarnames_list.append(similarnames)

    print(f"{name} -> {similarnames}")
```

```python
import ipywidgets as widgets
from IPython.display import display

output = widgets.Output()

dropdownobjects = []
for key, options in zip(names, similarnames_list):
    dropdownobjects.append(widgets.Dropdown(options=options, description=key))

input_widgets = widgets.VBox(dropdownobjects)
display(input_widgets)
```
